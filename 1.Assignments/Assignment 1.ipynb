*1. Write a python Program to display the last digit of number(hint: Any number divide by 10 will return last digit)*

*2. Write a function to Calculate Area and Perimeter of Rectangle*

```python
def calculate_rectangle_properties(length, width):
    area = length * width
    perimeter = 2 * (length + width)
    return area, perimeter

# Test the function
length = float(input("Enter the length of the rectangle: "))
width = float(input("Enter the width of the rectangle: "))

rectangle_area, rectangle_perimeter = calculate_rectangle_properties(length, width)

print("Area of the rectangle:", rectangle_area)
print("Perimeter of the rectangle:", rectangle_perimeter)
```

*3. Write a python Program to find out given No is Prime or Not.*

```python
def is_prime(number):
    if number <= 1:
        return False
    elif number <= 3:
        return True
    elif number % 2 == 0 or number % 3 == 0:
        return False
    i = 5
    while i * i <= number:
        if number % i == 0 or number % (i + 2) == 0:
            return False
        i += 6
    return True

# Test the function
num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a prime number.")
else:
    print(num, "is not a prime number.")
```

*4. Write a Python program that prompts the user to input three numbers and then determines and
prints the largest among them.*

```python
def find_largest_number(num1, num2, num3):
    return max(num1, num2, num3)

# Prompt the user to input three numbers
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
num3 = float(input("Enter the third number: "))

# Call the function to find the largest number
largest_number = find_largest_number(num1, num2, num3)

# Print the result
print("The largest number among", num1, ",", num2, ", and", num3, "is:", largest_number)
```

*5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)*

```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill_amount = 0
    elif units <= 200:
        bill_amount = (units - 100) * 5
    else:
        bill_amount = 100 * 5 + (units - 200) * 10
    return bill_amount

# Prompt the user to input the number of units
units = int(input("Enter the number of units consumed: "))

# Calculate the electricity bill
total_bill = calculate_electricity_bill(units)

# Print the total bill amount
print("Total bill amount:", total_bill)
```

*6. Write a program to accept numbers from 1-7 and display name of day like- 1 for sunday, 2 for Monday and so on.*

```python
def display_day_name(day_number):
    days = {
        1: "Sunday",
        2: "Monday",
        3: "Tuesday",
        4: "Wednesday",
        5: "Thursday",
        6: "Friday",
        7: "Saturday"
    }
    return days.get(day_number, "Invalid day number")

# Prompt the user to input a number from 1 to 7
day_number = int(input("Enter a number from 1 to 7: "))

# Display the name of the corresponding day
day_name = display_day_name(day_number)
print("The name of the day for number", day_number, "is:", day_name)
```

*7. Write a python program to check given word is Palindrome or not*

```python
def is_palindrome(word):
    # Convert the word to lowercase to handle case-insensitive comparison
    word = word.lower()
    # Check if the word is equal to its reverse
    return word == word[::-1]

# Prompt the user to input a word
word = input("Enter a word: ")

# Check if the word is a palindrome
if is_palindrome(word):
    print(word, "is a palindrome.")
else:
    print(word, "is not a palindrome.")
```

*8. Download the Dataset- https://www.kaggle.com/datasets/mohamedafsal007/house-price-dataset-of-india
Load the dataset.*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('House Price India.csv')

column_names = data.columns
print("Column Names:")
print(column_names)
```

*9. Perform Univariate Analysis*

```python
numerical_columns = ['number of bedrooms', 'number of bathrooms', 'living area','Built Year','Price']

# Visualize distributions of numerical columns
plt.figure(figsize=(18, 12))  # Increase the size of the figure

for i, col in enumerate(numerical_columns[:-1], 1):  # Exclude the last column (Price)
    plt.subplot(2, 2, i)  # Adjust the number of rows and columns as needed
    sns.histplot(data[col], kde=True)
    plt.title(col)
    plt.xlabel('')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

*10. Perform Bivariate Analysis*

```python
#Scatter Plot (Living Area vs. Price):

plt.figure(figsize=(10, 6))
sns.scatterplot(x='living area', y='Price', data=data)
plt.title('Living Area vs. Price')
plt.xlabel('Living Area')
plt.ylabel('Price')
plt.show()
```

```python
# Violin Plot (Number of Bedrooms vs. Price)
plt.figure(figsize=(10, 6))
sns.violinplot(x='number of bedrooms', y='Price', data=data)
plt.title('Number of Bedrooms vs. Price')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```
